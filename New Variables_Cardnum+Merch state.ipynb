```python
import pandas as pd
import numpy as np
import datetime as dt
```

### Amount variables (amount_cs)

```python
%%time
avg_cs_0=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='1D').mean().reset_index()
avg_cs_0['order_cs']=avg_cs_0.groupby(['Cardnum','Merch state','Date']).cumcount()+1
avg_cs_1=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='2D').mean().reset_index()
avg_cs_1['order_cs']=avg_cs_1.groupby(['Cardnum','Merch state','Date']).cumcount()+1
avg_cs_3=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='4D').mean().reset_index()
avg_cs_3['order_cs']=avg_cs_3.groupby(['Cardnum','Merch state','Date']).cumcount()+1
avg_cs_7=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='8D').mean().reset_index()
avg_cs_7['order_cs']=avg_cs_7.groupby(['Cardnum','Merch state','Date']).cumcount()+1
avg_cs_14=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='15D').mean().reset_index()
avg_cs_14['order_cs']=avg_cs_14.groupby(['Cardnum','Merch state','Date']).cumcount()+1
avg_cs_30=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='31D').mean().reset_index()
avg_cs_30['order_cs']=avg_cs_30.groupby(['Cardnum','Merch state','Date']).cumcount()+1

max_cs_0=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='1D').max().reset_index()
max_cs_0['order_cs']=avg_cs_0.groupby(['Cardnum','Merch state','Date']).cumcount()+1
max_cs_1=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='2D').max().reset_index()
max_cs_1['order_cs']=max_cs_1.groupby(['Cardnum','Merch state','Date']).cumcount()+1
max_cs_3=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='4D').max().reset_index()
max_cs_3['order_cs']=max_cs_3.groupby(['Cardnum','Merch state','Date']).cumcount()+1
max_cs_7=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='8D').max().reset_index()
max_cs_7['order_cs']=max_cs_7.groupby(['Cardnum','Merch state','Date']).cumcount()+1
max_cs_14=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='15D').max().reset_index()
max_cs_14['order_cs']=max_cs_14.groupby(['Cardnum','Merch state','Date']).cumcount()+1
max_cs_30=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='31D').max().reset_index()
max_cs_30['order_cs']=max_cs_30.groupby(['Cardnum','Merch state','Date']).cumcount()+1

med_cs_0=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='1D').median().reset_index()
med_cs_0['order_cs']=avg_cs_0.groupby(['Cardnum','Merch state','Date']).cumcount()+1
med_cs_1=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='2D').median().reset_index()
med_cs_1['order_cs']=med_cs_1.groupby(['Cardnum','Merch state','Date']).cumcount()+1
med_cs_3=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='4D').median().reset_index()
med_cs_3['order_cs']=med_cs_3.groupby(['Cardnum','Merch state','Date']).cumcount()+1
med_cs_7=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='8D').median().reset_index()
med_cs_7['order_cs']=med_cs_7.groupby(['Cardnum','Merch state','Date']).cumcount()+1
med_cs_14=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='15D').median().reset_index()
med_cs_14['order_cs']=med_cs_14.groupby(['Cardnum','Merch state','Date']).cumcount()+1
med_cs_30=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='31D').median().reset_index()
med_cs_30['order_cs']=med_cs_30.groupby(['Cardnum','Merch state','Date']).cumcount()+1

tot_cs_0=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='1D').sum().reset_index()
tot_cs_0['order_cs']=avg_cs_0.groupby(['Cardnum','Merch state','Date']).cumcount()+1
tot_cs_1=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='2D').sum().reset_index()
tot_cs_1['order_cs']=tot_cs_1.groupby(['Cardnum','Merch state','Date']).cumcount()+1
tot_cs_3=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='4D').sum().reset_index()
tot_cs_3['order_cs']=tot_cs_3.groupby(['Cardnum','Merch state','Date']).cumcount()+1
tot_cs_7=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='8D').sum().reset_index()
tot_cs_7['order_cs']=tot_cs_7.groupby(['Cardnum','Merch state','Date']).cumcount()+1
tot_cs_14=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='15D').sum().reset_index()
tot_cs_14['order_cs']=tot_cs_14.groupby(['Cardnum','Merch state','Date']).cumcount()+1
tot_cs_30=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='31D').sum().reset_index()
tot_cs_30['order_cs']=tot_cs_30.groupby(['Cardnum','Merch state','Date']).cumcount()+1
```

### Frequency Variables (freq_cs)

```python
%%time
freq_cs_0=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='1D').count().reset_index()
freq_cs_0['order_cs']=freq_cs_0.groupby(['Cardnum','Merch state','Date']).cumcount()+1
freq_cs_1=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='2D').count().reset_index()
freq_cs_1['order_cs']=freq_cs_1.groupby(['Cardnum','Merch state','Date']).cumcount()+1
freq_cs_3=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='4D').count().reset_index()
freq_cs_3['order_cs']=freq_cs_3.groupby(['Cardnum','Merch state','Date']).cumcount()+1
freq_cs_7=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='8D').count().reset_index()
freq_cs_7['order_cs']=freq_cs_7.groupby(['Cardnum','Merch state','Date']).cumcount()+1
freq_cs_14=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='15D').count().reset_index()
freq_cs_14['order_cs']=freq_cs_14.groupby(['Cardnum','Merch state','Date']).cumcount()+1
freq_cs_30=card_ref.groupby(['Cardnum','Merch state'])['Amount'].rolling(window='31D').count().reset_index()
freq_cs_30['order_cs']=freq_cs_30.groupby(['Cardnum','Merch state','Date']).cumcount()+1
```

```python
freq_cs=freq_cs_0.merge(freq_cs_1,on=['Cardnum','Merch state','Date','order_cs']).\
merge(freq_cs_3,on=['Cardnum','Merch state','Date','order_cs']).\
merge(freq_cs_7,on=['Cardnum','Merch state','Date','order_cs']).\
merge(freq_cs_14,on=['Cardnum','Merch state','Date','order_cs']).\
merge(freq_cs_30,on=['Cardnum','Merch state','Date','order_cs'])
```

```python
freq_cs.columns=['Cardnum','Merch state','Date','freq_cs_0','order_cs','freq_cs_1','freq_cs_3','freq_cs_7','freq_cs_14','freq_cs_30']
```

### Days Since Variable (diff_cs)

```python
card_sort=card_clean.sort_values(by=['Cardnum','Merch state','Date','order_cs'],ascending=True)
```

```python
card_sort2=card_sort.reset_index().iloc[:,1:]
```

```python
card_sort2['diff_cs']=0
```

```python
for i in range(len(card_sort2)-1):
    if card_sort2.loc[i,'Cardnum']==card_sort2.loc[i+1,'Cardnum'] and card_sort2.loc[i,'Merch state']==card_sort2.loc[i+1,'Merch state']:
        card_sort2.loc[i+1, 'diff_cs']=(card_sort2.loc[i+1,'Date']-card_sort2.loc[i,'Date']).days
    if card_sort2.loc[i,'Cardnum']!=card_sort2.loc[i+1,'Cardnum'] and card_sort2.loc[i,'Merch state']!=card_sort2.loc[i+1,'Merch state']:
        card_sort2.loc[i+1, 'diff_cs']=365
```

```python
card_sort2.loc[0, 'diff_cs']=365
```
